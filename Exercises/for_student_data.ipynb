# Make table of members for each club

```python
#rug = student_data['club'][student_data['club']=='Rugby']
club_data = {'Rugby': [], 
             'Improvisation': [],
             'Chess': [], 
             'Debate': [],
             'Football': []}
df = pd.DataFrame(club_data, columns = ['Rugby','Improv','Chess','Debate','Football'])
```

```python
Rugby = student_club['first_name'] [ (student_club[0]=='Rugby') | (student_club[1]==' Rugby') | 
                                      (student_club[2]==' Rugby') | (student_club[3]==' Rugby') ]
Improvisation = student_club['first_name'] [ (student_club[0]=='Improvisation') | (student_club[1]==' Improvisation') | 
                                      (student_club[2]==' Improvisation') | (student_club[3]==' Improvisation') ]
Chess = student_club['first_name'] [ (student_club[0]=='Chess') | (student_club[1]==' Chess') | 
                                      (student_club[2]==' Chess') | (student_club[3]==' Chess') ]
Debate = student_club['first_name'] [ (student_club[0]=='Debate') | (student_club[1]==' Debate') | 
                                      (student_club[2]==' Debate') | (student_club[3]==' Debate') ]
Football = student_club['first_name'] [ (student_club[0]=='Football') | (student_club[1]==' Football') | 
                                      (student_club[2]==' Football') | (student_club[3]==' Football') ]

club_data['Rugby'].append(Rugby)
club_data['Improvisation'].append(Improvisation)
club_data['Chess'].append(Chess)
club_data['Debate'].append(Debate)
club_data['Football'].append(Football)
```
